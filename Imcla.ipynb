Import necessary libraries

```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.utils import to_categorical
```

Load the CIFAR-10 dataset

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

Normalize the pixel values

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

Convert the labels to one-hot vectors

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

Define the neural network architecture

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3))) # Add convolutional layer with 32 filters, 3x3 kernel, ReLU activation, and input shape of 32x32x3
model.add(MaxPooling2D((2, 2))) # Add max pooling layer with 2x2 pool size
model.add(Conv2D(64, (3, 3), activation='relu')) # Add convolutional layer with 64 filters, 3x3 kernel, and ReLU activation
model.add(MaxPooling2D((2, 2))) # Add max pooling layer with 2x2 pool size
model.add(Conv2D(64, (3, 3), activation='relu')) # Add convolutional layer with 64 filters, 3x3 kernel, and ReLU activation
model.add(Flatten()) # Add flatten layer
model.add(Dense(64, activation='relu')) # Add dense layer with 64 units and ReLU activation
model.add(Dense(10, activation='softmax')) # Add dense layer with 10 units and softmax activation
```

Compile the model

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

Train the model

```python
model.fit(x_train, y_train, epochs=10, batch_size=64, validation_data=(x_test, y_test))
```

Evaluate the model on the test set

```python
loss, accuracy = model.evaluate(x_test, y_test)
print('Test loss:', loss) # Print test loss
print('Test accuracy:', accuracy) # Print test accuracy
```
